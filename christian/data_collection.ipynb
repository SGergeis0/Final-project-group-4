```python
import librosa
import numpy as np

def extract_features(christian/spotify_songs_nogenre.csv):
    y, sr = librosa.load(christian/spotify_songs_nogenre.csv)
    mfccs = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(y=y, sr=sr).T, axis=0)
    tempo = librosa.beat.tempo(y, sr=sr)[0]
    return np.hstack([mfccs, chroma, tempo])

# Apply feature extraction to both datasets (pseudo-code for the process)
#gtzan_data['features'] = gtzan_data['file_path'].apply(extract_features)
spotify_data['features'] = spotify_data['christian/spotify_songs_nogenre.csv'].apply(extract_features)
```
